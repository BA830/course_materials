```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

from pyfixest.estimation import feols
from pyfixest.utils import get_data
from pyfixest.summarize import etable
```

```python
df_titanic = pd.read_csv("../data/titanic.csv")
df_titanic.head()
```

```python

# Convert to categorical:
df_titanic['Pclass_cat'] = df_titanic['Pclass'].astype('category')

model_nofe = feols("Survived ~  Age + Pclass_cat", data = df_titanic).vcov("hetero")
model = feols("Survived ~  Age | Pclass", data = df_titanic).vcov("hetero")
etable([model_nofe, model])
```

```python
# Can put in many fixed effects:
model = feols("Survived ~  Age | Pclass + Embarked", data = df_titanic).vcov("hetero")
etable(model)
```
